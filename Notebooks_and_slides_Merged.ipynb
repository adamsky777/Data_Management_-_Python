## Variables

Variables are essential in (almost) any programming language: they are used to label and store values in memory (in order to use these values later).

In Python, a new variable is created by assignment:
```python
variable_name = some_value
```
For example (note that each statement is on a separate line):

````python
width = 3.5   # Width of a rectangle (anything after a # on the same line is a comment; unless a part of a string)
height = 2.5  # Height of a rectangle
area = width * height
print("Area of a rectangle with width", width, "and height", height, "is", area)
````

Using an undefined variable will lead to an error:

````python
width * heigth  # Note the typo in height
````

Variable names (also known as identifiers) must only contain underscores, letters and digits. They cannot start with a digit and they cannot be a keyword (reserved words that have a special meaning). Variable names are case-sensitive, i.e. `area` and `Area` denote two different variables.

**A common practice is to use *snake_case* for variable names, i.e. lower case, words separated by an underscore, for example `total_area`.**

Tip: Always use nouns that describe the purpose of variables so that it is easy for other people (and you) to read your code.

````python
# None of these is a legal variable name:
@name = 1                          # variable name contains an illegal character
1st_participant = 'Homer Simpson'  # variable name cannot start with a digit
from = '2021-06-21'                # "from" is a keyword
````

Avoid using the names of existing built-in or imported functions and types, since this will make it impossible to use them:

## Introduction to lists

**A list is an ordered collection of values**. Lists are often called arrays in other programming languages. Note that in Python, the individual elements might be of different types.

A list is created by specifying its elements, separated by commas, between square brackets (`[` and `]`):

````python
lst1 = []  # Empty list
lst2 = [1, 2, 3]
lst3 = [1.2]
lst4 = [["A", "B"], True]  # Elements can be of any types, even lists (or dictionaries that we will cover later)
lst5 = [2, False, 3.14, "Hello"]
````

We can use the `len` function to return the number of elements:

````python
lst = [2, False, 3.14, "Hello"]
len(lst)
````

Similar to strings (which are sequences of characters), we can use the index `[]` and slice `[:]` operators to access individual elements and slices:

````python
lst = ['A', 'B', 'C', 'D']
# Exercise: can we name the variable `list` instead of `lst`?

print(lst[0])    # First element
print(lst[-1])   # Last element,
print(lst[2:4])  # A slice between index 2 (inclusive) and 4 (exclusive)
````

Unlike strings, which are immutable, **lists are mutable** so we can change the individual elements:

````python
lst = ['A', 'B', 'C', 'D']
lst[0] = 'E'  # Change the first element
lst[-2] = 'F' # Change the next-to-last element
print(lst)
````

We can also update a whole slice at once:

````python
lst = ['A', 'B', 'C', 'D']
lst[1:3] = ['E', 'F']  # Elements B and C will be replaced by E and F
print(lst)
````

When updating a slice, the new value can have the size different from the size of the slice:

````python
lst = ['A', 'B', 'C', 'D']
lst[1:3] = ['E', 'F', 'G']  # Elements B and C will be replaced by E, F and G
print(lst)
````

````python
lst = ['A', 'B', 'C', 'D']
lst[1:3] = []  # Elements B and C will be removed.
print(lst)
````

Updating slices can be used to append new elements at the end of a list as well:

````python
lst = ['A', 'B', 'C', 'D']
lst[len(lst):] = ['E', 'F']
print(lst)
````

We will later return to lists and see other (and more ergonomic) ways how to insert and remove elements to and from them, but let us already now mention `append`, one of the most useful list methods that adds a (single) element to the end of a list:

````python
lst = ['A', 'B', 'C', 'D']
lst.append('E')
print(lst)
````

## Conditional execution

So far we have seen examples of Python code where all lines (each containing one statement) were executed sequentially. What if we need to execute a block of statements only when some given condition is true? We can use an `if` statement:

```python
if condition:
    # ...
    # indented block of code that will be executed only if the condition is true
    # ...
```

Python does not use special characters to denote the beginning and the end of a block (like `{` and `}` in C/C++/Java) but uses *indentation*, i.e. the amount of white spaces at the beginning of each line.

All lines in the same block must have the same indentation. The indentation of a block that is going to be executed when an `if` condition is true must be larger than the indentation of the `if` line. The Python style guide recommends to increase the indentation by 4 spaces (and not use use tabs).

Note that there must be at least one statement in the block. If you don't want to execute any code if the condition is met (e.g. the real code will be added later), use `pass` (a statement that does not do anything):

```python
if condition:
    pass  # We will add the code later.
```

Following example asks the user for an integer number and then checks if it is less than 0. If so, it will print that it is a negative number. Try running the cell multiple times and try to enter both negative and non-negative values.

````python
x = int(input("Enter an integer number:"))
if x < 0:
    print("That's a negative number.")
    print("Yes, it is not positive, and not a zero either!")
print("Done!")  # This will run either x < 0 or not.
````

We can also specify a block of statements to be executed if the condition is not true (using `else`):

```python
if condition:
    # executed if the condition is true
else:
    # executed if the condition is false
```

Conditionals can be chained (using `elif` and `else`):
```python
if condition_1:
    # executed if condition_1 is true
elif condition_2:
    # executed if condition_1 is false, but condition_2 is true
elif condition_3:
    # executed if condition_1 and condition_2 are both false, but condition_3 is true
else:
    # executed if all of the conditions are false
```

It is also possible to nest conditionals:
```python
if condition_1:
    # ...
    if condition_2:
        # executed if both condition_1 and condition_2 are true
    else:
        # executed if condition_1 is true but condition_2 is false
    # executed if condition 1 is true
else:
    # ...
```

### For loops

`For` loops iterate over (i.e. go through) all elements of a list, a range of numbers, or, in general, a value of a type that is capable of returning its elements one by one (such types are called *iterables*), and execute a given block of statements for each element. In this block, the value of the current element is stored in the provided variable name:

```python
for variable in iterable:
    # ...
    # indented block of statements
    # (usually doing something with `variable` which stores the value of the current element)
    # ...
```

For example:

````python
for person in ['Homer', 'Marge', 'Bart', 'Lisa', 'Maggie']:
    print(person)
print('Done')
````

Above we have seen an example how to use a `while` loop to iterate over integer numbers in a given interval. More common way how to achieve this is to use a `for` loop iterating over a `range`:

````python
for i in range(10):
    print(i)
````

The `range(stop)` function returns a range of numbers from 0 (inclusive) to *stop* (exclusive!). If we want to start from a different *start* value, we can use `range(start, stop)`. (There is also a parameter to specify the step increment, look it up in the documentation when/if you need it.)

````python
for i in range(3, 10):  # In Matlab `for i = 3:9`, in R `for (i in 3:9)`
    print(i)
````

Strings are iterable, so you can iterate over their characters in a loop:

````python
for character in "Hello, world!":
    print(character)
````

**NEVER** change the list (iterable) you are iterating over inside of the loop! For example, never do the following:

````python
alist = [1, 2, 3, 4]
for element in alist:
    print(element)
    alist.remove(element)  # NEVER DO THIS!!!
print('Done, the list is', alist)
````

### Continue and break

We can use `continue` to interrupt the current iteration and continue with the next one. The following example prints number from 1 to 10 that are not divisible by 2 or 3:

````python
i = 0
while i < 10:
    i += 1
    if i % 2 == 0 or i % 3 == 0:
        continue
    print(i)
print("Done")
````

**In-class exercise.** What would happen if we moved `i += 1` after the `if` statement?

We can also interrupt the whole loop with `break`. The following example keeps printing each character of a given string on a separate line until the first occurrence of `!`.

````python
text = "Hello, dude! How are you doing?"

for character in text:
    if character == '!':
        break
    print(character)
print("Done")
````

Sometimes it might be convenient to create an infinite loop and use `break` to end it:

````python
rate = 10.4
while True:
    amount_str = input("Amount in NOK (or press Enter to exit): ")
    if amount_str == "":
        break
    amount = float(amount_str)
    amount_eur = amount / rate
    print("{:.2f} NOK = {:.2f} EUR".format(amount, amount_eur))
print("Thanks for using this tool ;)")
````

## Lists

We briefly introduced lists in the first notebook, and now we will look at them in more detail.

Recall that we can use the slice operator to add and remove elements from lists, but it is might be easier to use theirs methods:

````python
lst = ['A', 'B', 'C', 'D']
lst.insert(1, 'E')      # Adds 'E' at index 1 (specified by the first argument)
lst.append('F')         # Adds 'F' to the end (`append`s a single element to the end)
lst.extend(['G', 'H'])  # Extends the list with another list)
print(lst)

lst = ['A', 'B', 'C', 'D']
del lst[2]              # Removes the element at index 2 (C)
print(lst)

lst = ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D']
lst.remove('B')         # Removes the first occurence of 'B'
print(lst)

lst = ['A', 'B', 'C', 'D']
element = lst.pop()     # Removes the last element and returns it
print(lst, " and the removed element is:", element)
````

## Tuples

A tuple is created by specifying its elements, separated by commas, between parentheses (`(` and `)`):

````python
atuple = ()    # Empty tuple (we could also use `atuple = tuple()`)
atuple = (1,)  # A tuple with one element. **NOTE THE TRAILING COMMA!!!**
atuple = ('A', 'B', 'C', 'D')
````

Notes:
* In the example above we could omit the parentheses, i.e. we could write `atuple = 'A', 'B', 'C', 'D'`. (We cannot always do that, consider e.g. using a tuple as a parameter in a function or as an element of a list.)
* We cannot use e.g. `(1)` to create a tuple with a single element: `(1)` has the `int` type&mdash;it's just just an integer number in parentheses. To create a tuple with one element we need to include a trailing comma after the element, e.g. `(1,)` (or `1,`).

````python
print("Type of (1):", type((1)))
print("Type of (1,):", type((1,)))
````

The expression on the right side (`y, x`) is actually a tuple, and we are assigning its first element to `x` and its second element to `y`.

## Dictionaries

**A dictionary is a collection of key-value pairs.** In other languages, dictionaries might be called associative arrays, maps or symbol tables.

Each key can appear in the dictionary only once. The most common type for keys is string (*str*), but you can use any other immutable type as well, including tuples. We will however only use strings as keys.

Values for different keys can be of different types (even composite types such as lists, tuples and dictionaries). 

A dictionary is created by specifying the *key: value* pairs, separated by commas, between curly brackets (`{` and `}`):

````python
employee = {
    "name": "Alice",
    "surname": "Newton",
    "year_of_birth": 1970,
    "email": ["newton@company.com", "a.newton@company.com"]
}
````

````python
for key, value in employee.items():
    print(key, value)
````

## Syntax errors

If we try to execute a program that is not syntactically correct, Python will not run anything and will instead print a syntax error message:

## Exceptions

If a program is syntactically correct, but during its execution does something what is not allowed, Python will generate (*raise*) an *exception*. By default, any exception leads to stopping the execution of the program and showing an error message to the user.

We have already seen quite a few types of exceptions:

````python
1 / 0
````

````python
a = [1, 2, 3]
a[5]
````

````python
1 + "2"
````

````python
int("Hello!")
````

We can, however, change the default behavior by *catching* exceptions, and specifying how to handle them with the help of a *try* and *except* block:
```python
try:
    #
    # Code to execute
    #
except:
    # 
    # Code that will be executed (only) if any exception is raised during the `try` block
    #
```
Note that if an exception is raised during the execution of the `try` block, the remaining code in the block is not executed. 

For example:

````python
string = "Hello"
try:
    value = int(string)
    print("The integer value is {}".format(value))  # This will not be executed because of an exception
except:
    print('Something bad happened')
````

If we want to handle only a certain type of exceptions, we can specify the type (the class name) after `except`:

````python
string = "Hello"
try:
    value = int(string)
    print("The integer value is {}".format(value))  # This will not be executed because of an exception
except ValueError:
    print('Could not convert the given string to an integer!')
````

To store the caught exception in a variable, we can add `as` and the variable name after the type:

````python
string = "Hello"
try:
    value = int(string)
except ValueError as e:
    print('Error:', e)
    print("Using 0 instead!")
    value = 0
print("The value is {}".format(value))
````

If we want to catch any exception and store it in a variable, we can use the type `Exception` as shown in the following example:

````python
string = "Hello"
try:
    value = int(string)
    print("The integer value is {}".format(value))  # This will not be executed because of an exception
except Exception as e:
    print('Something bad happened, see the error message below:')
    print(e)
````

Notes:
- If you want to use the same `except` block for handling errors of several different types, specify them in the following way:
    ```python
    try:  
       # Code
    except (OperationalError, InternalError):
       # Handling of an exception
    ```
- Use several `except` blocks if you want to handle different types of exceptions differently:
    ```python
    try:  
       # Code
    except OperationalError:
       # Handling of an OperationalError
    except InternalError:
       # Handling of an InternalError
    ```
    

If there is code that needs to be executed either there was an exception or not, we can use a `finally` block:

````python
string = "Hello"
try:
    value = int(string)
    print("The integer value is {}".format(value))  # This will not be executed because of an exception
except ValueError as e:  # Note that we have now specified the exception type!
    print('Something bad happened, see the error message below:')
    print(e)
finally:
    print("Ok, we are done here...")
````

## Monte Carlo method with NumPy

One of the most known examples of the Monte Carlo method is estimating $\pi$ by placing a "huge" number of random points in a square with the side length of 1 (red and blue points in the figure below), and calculating the ratio of those that are within a quarter circle of radius 1 (red points). Given that the area of the quarter circle is $\pi/4$ and the area of the square is $1$, we can estimate $\pi$ by taking this ratio and multiplying it by 4.

Let us try to do this experiment using random arrays in NumPy:

````python
N = 1_000_000  # We will place 1 mill. points in the square
x = npr.random(N)  # The x-coordinates of N points
y = npr.random(N)  # The y-coordinates of N points
inside_of_unit_circle = (x**2 + y**2 <= 1)   # Are they inside of the unit circle?
pi_est = 4 * inside_of_unit_circle.sum() / N
print("Estimate of pi:", pi_est)
````

## Series

A Pandas series is a 1-dimensional array with an *index*—a set of labels associated with the values.

## Combining data frames

Consider the following example. We have two data frames, one containing data about the height of a group of people, the second about the weight:

````python
height_df = pd.DataFrame([181, 167, 200], index=["A", "B", "C"], columns=["height"])
weight_df = pd.DataFrame([72.3, 67.0, 82.3], index=["A", "B", "D"], columns=["weight"])
display(height_df)
display(weight_df)
````

You might have noticed that the first data frame does not contain the height for *D* and the second data does not contain the weight for *C*.

We can combine these to data frames using the `merge` method:

````python
pd.merge(height_df, weight_df, left_index=True, right_index=True)
# The first parameter is called the left frame, the second the right frame.
# We are using the labels of the left frame (left_index=True) to match with
# the labels of the right frame (right_frame=True).
````

The resulting data set contains rows only for people for which we know both the height and the weight. Such a result is called an *inner join*.

If we want to include all people that exist in the *left* frame (i.e. `height_df`), even if no matching rows are found in the right frame, we can do an *(outer) left join*:

````python
pd.merge(height_df, weight_df, how="left", left_index=True, right_index=True)
````

Note the NaN (representing the missing value) in the weight column for person *C*.

We can also include all people in the *right* frame, even if no matching rows are found in the left frame with an *(outer) right join*:

````python
pd.merge(height_df, weight_df, how="right", left_index=True, right_index=True)
````

Finally, we can do a *full outer join* where people from both the left and right frame will be included:

````python
pd.merge(height_df, weight_df, how="outer", left_index=True, right_index=True)
````

Note: If we want to match rows based on the data in columns instead of the indexes, we can specify the column names using the `left_on` and `right_on` parameters instead of `left_index=True` and `right_index=True`. If the columns have the same name, we can just specify the name using the `on` parameter.

````python
df.fillna(0.0)
````

We can use the same method to replace the missing data with the summary statistics based on the non-null values in each column, e.g. the mean:

To deal with wrong data, we can use the `replace` method:

````python
df['C1'].replace('E', 'F')
````

## Prepared statements

To run SQL statements that use potentially unsafe data (e.g. data entered by users) we can (and **should**!) use prepared statements. A prepared statement is an SQL that uses placeholders (`?` in sqlite3) that will be replaced by the specified values in a secure way. (Note that other packages might use different placeholders, e.g. `%s`, more information [here](https://www.python.org/dev/peps/pep-0249/#paramstyle).)

For example, if we want to find employees born between two given years, and these are entered by user, we can use the following prepared statement:
```sql
SELECT * FROM employee WHERE year_of_birth BETWEEN ? AND ?
```
and specify a tuple with the values (that will replace the placeholders) as the second parameter when calling the `execute` method (remember a trailing comma if there is only one parameter in the tuple):

````python
year_from = 1975
year_to = 1980

cursor.execute("SELECT * FROM employee WHERE year_of_birth BETWEEN ? AND ?", (year_from, year_to))

for row in cursor:
    print(row)
````

## SQL injection

Why is it so important to use prepared statements? Consider the following program:

````python
employee_id = input('Enter employee ID: ')

cursor.execute("SELECT * FROM employee WHERE id = {}".format(employee_id))

for row in cursor:
    print(row)
````

This program does exactly what we want, as long we enter valid employee IDs. But what happens if you enter the following input?
```

```

The query that gets executed is actually:
```sql
SELECT * FROM employee WHERE id = 0
UNION
SELECT *, NULL, NULL, NULL, NULL, NULL FROM project
```
returning the following rows:
```
(1, 'Project Alfa, Uppsala', 210000, None, None, None, None, None)
(2, 'Project Bravo, Knivsta', 190000, None, None, None, None, None)
(3, 'Project Charlie, Stockholm', 420000, None, None, None, None, None)
```

If executing multiple statements were enabled (and they are in some database packages), we would be able to do even more damage:

Before we start creating any tables, we need to think about what information we need to store in the database, and how to divide it into tables. For each table we need to answer the following questions:

- What is the table name?
- What attributes (columns) are we going to store in the table? For each attribute we need to choose:
  - its name
  - its data type (see below)
  - constraints we wish to enforce, we can for example require that:
    - NULLs are disallowed (by default, NULLs are allowed)
    - all values must be unique
- What attribute(s) will be the primary key (and uniquely identify each row in the table)?
- What attributes are foreign keys (and what are they referring to)?

Recall the most common data types used in relational databases:

- BOOLEAN
- INT
- FLOAT
- DOUBLE
- DECIMAL($p$, $s$), exact numeric value with $p$ significant digits and $s$ digits after the decimal point.
- CHAR($l$), fixed-length string of length $l$. (Max. value for $l$ is many RDBMS is 255.)
- VARCHAR($l$), variable-length string of max. length $l$. (Max. value for $l$ in many RDBMS is 255.)
- ENUM($s_1, s_2, \dots, s_n$), value from a set of allowed strings. (Note: There is no ENUM in SQLite, but you can assign an integer code to each allowed string and use INT).
- TEXT, suitable for texts that might be longer than 255 characters.
- DATE
- DATETIME

Let us create a database to store the final grades of students in different courses at BI. For pedagogical reasons we will keep the database as simple as possible, only storing a few columns:

- In the **student** table we will store the following attributes (columns) for each student:
  - *id*: integer identifier, mandatory (i.e. NULLs are not allowed), unique for each student,
  - *first_name*: string, max. 255 characters, mandatory,
  - *last_name*: string, max. 255 characters, mandatory.
  
  We will use *id* as the primary key.


- In the **course** table we will store the following attributes for each course: 
  - *id*: integer identifier, mandatory and unique for each course,
  - *code*: string, max. 30 characters, mandatory and also unique,
  - *title*: string, max. 255 characters, optional (we choose optional for demonstration purposes, in the real database we would require the attribute to be mandatory).
    
  We will use *id* as the primary key (we could also use *code*).


- In the **grade** table we will store the grades:
  - *id_student*: ID of the student,
  - *id_course*: ID of the course,
  - *grade*: a single character, mandatory. The is the grade that the student identified by *id_student* received in the course identified by *id_course*.

  Both *id_student* and *id_course* are foreign keys, the former references the primary key (*id*) in the *student* table, and the latter the primary key (*id*) in the *course* table. Since we store only the final grade, a combination of the *id_student* and *id_course* uniquely identifies each row and we will use it as the primary key.

We are now ready to create the tables:

````python
%%sql_script
CREATE TABLE student(
  id INT NOT NULL PRIMARY KEY,
  first_name VARCHAR(255) NOT NULL,
  last_name VARCHAR(255) NOT NULL
);
CREATE TABLE course(
  id INT NOT NULL PRIMARY KEY,
  code VARCHAR(30) NOT NULL UNIQUE,
  title VARCHAR(255)
);
CREATE TABLE grade(
  student_id INT NOT NULL,
  course_id INT NOT NULL,
  grade CHAR(1) NOT NULL,
  PRIMARY KEY(student_id, course_id),
  FOREIGN KEY(student_id) REFERENCES student(id),
  FOREIGN KEY(course_id) REFERENCES course(id)
);
````

Pay attention to how we specified the columns, their types and constraints, including the primary and the foreign keys.

Note that in the tables *student* and *course* that use a single attribute as the primary key, we included `PRIMARY KEY` at the end of the specification for *id*. However, if the primary key contains several attributes, we need to specify it as shown in the `CREATE TABLE` statement for the *grade* table.

Homer was indeed removed from the student table so were all the tuples from the grade table referencing the removed student.

The ``ON DELETE CASCADE`` specification says that all grade rows for a particular student are going to be removed when that student is removed. (And similarly, all grade rows for a particular course are going to be removed when that course is removed because of the second ``ON DELETE CASCADE``.)

Very similar situation occurs if we try to change the primary key for a student. If we want to change the ID of Marge Simpson to, say, 100 (nobody has this ID so we are allowed to do that):

````python
%%sql_script
UPDATE student SET id=100 WHERE id=2
````

We will get an error as well (Error: ``FOREIGN KEY constraint failed``). The reason is that the grade table has a row saying "student with ID 2" got grade of 5 in "course with ID 1" and after changing the ID there would be no student with ID 1. We can force updating student IDs in other tables by adding ``ON UPDATE CASCADE``:

````python
%%sql_script
-- Drop the table:
DROP TABLE grade;
-- Create the table again:
CREATE TABLE grade(
  student_id int NOT NULL,
  course_id int NOT NULL,
  grade char(1) NOT NULL,
  PRIMARY KEY(student_id, course_id),
  FOREIGN KEY(student_id) REFERENCES student(id) ON DELETE CASCADE ON UPDATE CASCADE,
  FOREIGN KEY(course_id) REFERENCES course(id) ON DELETE CASCADE ON UPDATE CASCADE
);
-- Insert the original content into the table:
INSERT INTO grade(student_id, course_id, grade) VALUES (2, 1, 'A');
````
