### Assignment 1.2 (2 points)

A *PSN* number is a non-negative integer number that is equal to the sum of its digits multiplied by the product of its digits. For example, 135 is a PSN number since $(1 + 3 + 5) * (1 * 3 * 5) = 135$, but 27 is not since $(2 + 7) * (2 * 7) = 126 \not= 27$.

1. Write a function `is_psn` that returns `True` if the argument is a PSN number, and `False` otherwise.

   **Sample arguments and return values**

        Argument: 135
        Returns: True
    
        Argument: 27
        Returns: False

2. Use the function `is_psn` function to find and print all PSN numbers less than 1000.

````python
# ANSWER CELL 1.2 (DO NOT REMOVE OR CHANGE THIS LINE)
import numpy as np
# 1.
def is_psn(number):
    pass
    # Your code here
    n = str(number)
    numbers = [int(i) for i in n]
    numbers = ((sum(numbers)) * np.prod(numbers))
    return (numbers == number)

print("Test 1 passed:", is_psn(135) == True)
print("Test 1 passed:", is_psn(27) == False)

# 2.


m = list(range(0, 1000))
print("\n part 2 \n"
      "PSN numbers between 0-1000:")
for i in m:
    if is_psn(i) == True:
        print(i)
````

Write a function `var` that accepts a non-empty list of floating-point numbers and calculates their variance.

The variance of $\{x_i\}_{i=1}^N$ is defined as
$$ \frac{1}{N} \sum_{i=1}^N (x_i - \overline x)^2, $$
where
$$ \overline x = \frac{1}{N} \sum_{i=1}^N x_i $$

**You are NOT allowed to use any imported packages and their functions/methods in your implementation!**

````python
# ANSWER CELL 1.3 (DO NOT REMOVE OR CHANGE THIS LINE)

empty_list = []
def var(lst):
    pass
    print(lst)
    if lst != []:
        N = len(lst)
        avg = (sum(lst))/N
        for i in lst:
            x = (i-avg)
            x = x**2
            empty_list.append(x)

    return ((sum(empty_list))/N)





for i in range(1, 6):
    lst = 10 * np.random.random(size=5)
    print("Test", i, "passed:", abs(var(lst.tolist()) - lst.var()) < 1e-5)
````

Write a function `stats` that accepts a list of strings (with English words) and returns a dictionary with the following keys and values:

- `"empty"`: the number of empty strings in the list (or 0 if there are no such strings),
- `"short"`: the number of non-empty strings that have 5 or fewer characters (or 0 if there are no such strings),
- `"long"`: the number of strings that have more than 5 characters (or 0 if there are no such strings).

**Sample arguments and return values**

    Argument: ["enclose", "mapped", "the", "untunes", "excavator", "", "conserves", "brown", "knee"]
    Returns: {'empty': 1, 'short': 3, 'long': 5}

    Argument: []
    Returns: {'empty': 0, 'short': 0, 'long': 0}

````python
# ANSWER CELL 1.4 (DO NOT REMOVE OR CHANGE THIS LINE)

def stats(strings):
    pass
    empty = 0
    short = 0
    long = 0
    for i in strings:
        if len(i) < 1:
            empty += 1
        elif 6> len(i) >=1:
            short +=1
        elif len(i)>=6:
            long +=1
    moby_dict = {'empty': empty, 'short': short, 'long': long}
    return moby_dict


print("Test 1 passed:",
      stats(["enclose", "mapped", "the", "untunes", "excavator", "", "conserves", "brown", "knee"]) ==
      {'empty': 1, 'short': 3, 'long': 5})
print("Test 2 passed:", stats([]) == {'empty': 0, 'short': 0, 'long': 0})
````

## Appendix A. Database schema

```sql
CREATE TABLE Categories
(
    CategoryID INTEGER PRIMARY KEY,
    CategoryName TEXT,
    Description TEXT
);

CREATE TABLE Customers
(
    CustomerID INTEGER PRIMARY KEY,
    CustomerName TEXT,
    ContactName TEXT,
    Address TEXT,
    City TEXT,
    PostalCode TEXT,
    Country TEXT
);

CREATE TABLE Employees
(
    EmployeeID INTEGER PRIMARY KEY,
    LastName TEXT,
    FirstName TEXT,
    BirthDate DATE,
    Photo TEXT,
    Notes TEXT
);

CREATE TABLE Shippers(
    ShipperID INTEGER PRIMARY KEY,
    ShipperName TEXT,
    Phone TEXT
);

CREATE TABLE Suppliers(
    SupplierID INTEGER PRIMARY KEY,
    SupplierName TEXT,
    ContactName TEXT,
    Address TEXT,
    City TEXT,
    PostalCode TEXT,
    Country TEXT,
    Phone TEXT
);

CREATE TABLE Products(
    ProductID INTEGER PRIMARY KEY,
    ProductName TEXT,
    SupplierID INTEGER,
    CategoryID INTEGER,
    Unit TEXT,
    Price NUMERIC DEFAULT 0,
    FOREIGN KEY (CategoryID) REFERENCES Categories (CategoryID),
    FOREIGN KEY (SupplierID) REFERENCES Suppliers (SupplierID)
);

CREATE TABLE Orders(
    OrderID INTEGER PRIMARY KEY,
    CustomerID INTEGER,
    EmployeeID INTEGER,
    OrderDate DATETIME,
    ShipperID INTEGER,
    FOREIGN KEY (EmployeeID) REFERENCES Employees (EmployeeID),
    FOREIGN KEY (CustomerID) REFERENCES Customers (CustomerID),
    FOREIGN KEY (ShipperID) REFERENCES Shippers (ShipperID)
);

CREATE TABLE OrderDetails(
    OrderDetailID INTEGER PRIMARY KEY,
    OrderID INTEGER,
    ProductID INTEGER,
    Quantity INTEGER,
```
